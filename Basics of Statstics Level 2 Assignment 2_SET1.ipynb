# Finding the Outlier 

```python
sns.boxplot(data["Measure_X(in %)"])
plt.grid()
```

# In above box plot we see that there is ONE outliers one is "Morgan Stanley(91.36)"

```python
#Using pie chart
plt.figure(figsize=(10,10))
plt.pie(data["Measure_X(in %)"],labels=data["Name of company"],shadow = True,autopct='%1.2f%%')
plt.title("Various measures corresponding to the company")
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data["Name of company"],data["Measure_X(in %)"],orient="v")
plt.xticks(rotation = 90, fontsize = 12, fontweight = 'bold')
for i in range(len(data["Name of company"])):
    plt.text(i,data["Measure_X(in %)"][i],data["Measure_X(in %)"][i],ha="center",va="bottom")
```
